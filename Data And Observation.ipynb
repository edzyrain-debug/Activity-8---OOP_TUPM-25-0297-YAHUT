```python
# 3
class BankAccount: 
    def __init__(self, account_holder, balance): 
        self.__balance = balance  # Private attribute 
        self.account_holder = account_holder 
 
    def deposit(self, amount): 
        self.__balance += amount 
        return f"New Balance: {self.__balance}" 
 
    def get_balance(self): 
        return f"Balance: {self.__balance}"
account = BankAccount("John Doe", 1000) 
print(account.deposit(500)) 
print(account.get_balance())
```

```python
from abc import ABC, abstractmethod 

class Appliance(ABC): 
    @abstractmethod 
    def turn_on(self): 
        pass 

class Fan(Appliance): 
    def turn_on(self): 
        return "Fan is spinning"

fan = Fan() 
print(fan.turn_on())
```

```python
# 14
class ATM: 
    def __init__(self, balance): 
        self.balance = balance 

    def withdraw(self, amount): 
        if amount > self.balance: 
            return "Insufficient balance" 
        self.balance -= amount 
        return f"Withdrawn {amount}, Remaining balance: {self.balance}"

atm = ATM(1000) 
print(atm.withdraw(200))
```

```python
# Library Book Management System

books = []

# Input number of books
while True:
    try:
        num_books = int(input("Enter number of books to add: "))
        if num_books <= 0:
            print("Please enter at least 1 book.")
        else:
            break
    except ValueError:
        print("Invalid input. Enter a valid number.")

# Add books
for i in range(num_books):
    print(f"\nBook {i+1}")

    title = input("Enter title: ")
    author = input("Enter author: ")

    books.append({
        "title": title,
        "author": author,
        "available": True
    })

# Display books
def display_books():
    print("\nLibrary Books:")
    for book in books:
        status = "Available" if book["available"] else "Borrowed"
        print(f"{book['title']} by {book['author']} - {status}")

# Borrow book
borrow_title = input("\nEnter book title to borrow: ")

found = False
for book in books:
    if book["title"].lower() == borrow_title.lower():
        found = True
        if book["available"]:
            book["available"] = False
            print(f"{book['title']} has been borrowed.")
        else:
            print("Book is already borrowed.")
        break

if not found:
    print("Book does not exist.")

# Return book
return_title = input("\nEnter book title to return: ")

found = False
for book in books:
    if book["title"].lower() == return_title.lower():
        found = True
        if not book["available"]:
            book["available"] = True
            print(f"{book['title']} has been returned.")
        else:
            print("Book was not borrowed.")
        break

if not found:
    print("Book does not exist.")

# Final display
display_books()
```

```python
# Employee Payroll System

employees = []

# Input number of employees
while True:
    try:
        num_employees = int(input("Enter number of employees: "))
        if num_employees <= 0:
            print("Please enter at least 1 employee.")
        else:
            break
    except ValueError:
        print("Invalid input.")

# Employee data input
for i in range(num_employees):
    print(f"\nEmployee {i+1}")

    name = input("Enter employee name: ")

    while True:
        try:
            hourly_rate = float(input("Enter hourly wage: "))
            if hourly_rate < 0:
                print("Hourly wage cannot be negative.")
            else:
                break
        except ValueError:
            print("Invalid input.")

    while True:
        try:
            hours_worked = float(input("Enter total hours worked: "))
            if hours_worked < 0:
                print("Hours worked cannot be negative.")
            else:
                break
        except ValueError:
            print("Invalid input.")

    # Salary calculation
    if hours_worked > 160:
        overtime_hours = hours_worked - 160
        base_salary = 160 * hourly_rate
        overtime_pay = overtime_hours * (hourly_rate * 1.5)
    else:
        overtime_hours = 0
        base_salary = hours_worked * hourly_rate
        overtime_pay = 0

    total_salary = base_salary + overtime_pay

    employees.append({
        "name": name,
        "hours": hours_worked,
        "base_salary": base_salary,
        "overtime_pay": overtime_pay,
        "total_salary": total_salary
    })

# Display payroll
print("\nEmployee Payroll Summary")
for emp in employees:
    print(f"""
Name: {emp['name']}
Hours Worked: {emp['hours']}
Base Salary: ₱{emp['base_salary']:.2f}
Overtime Pay: ₱{emp['overtime_pay']:.2f}
Total Salary: ₱{emp['total_salary']:.2f}
""")
```

```python
# Online Shopping Cart System

products = {
    "Laptop": {"price": 50000, "stock": 5},
    "Mouse": {"price": 500, "stock": 10},
    "Keyboard": {"price": 1500, "stock": 7}
}

cart = {}

# Display products
def display_products():
    print("\nAvailable Products:")
    for product, details in products.items():
        print(f"{product} - ₱{details['price']} | Stock: {details['stock']}")

# Add to cart
while True:
    display_products()

    choice = input("\nEnter product to add (or 'done' to finish): ")

    if choice.lower() == "done":
        break

    if choice not in products:
        print("Invalid product selection.")
        continue

    try:
        quantity = int(input("Enter quantity: "))

        if quantity <= 0:
            print("Quantity must be positive.")
            continue

        if quantity > products[choice]["stock"]:
            print("Not enough stock available.")
            continue

        # Add item
        products[choice]["stock"] -= quantity

        if choice in cart:
            cart[choice]["quantity"] += quantity
        else:
            cart[choice] = {
                "price": products[choice]["price"],
                "quantity": quantity
            }

        print(f"{choice} added to cart.")

    except ValueError:
        print("Invalid quantity.")

# Remove item
remove_item = input("\nEnter product to remove from cart: ")

if remove_item in cart:
    removed_qty = cart[remove_item]["quantity"]
    products[remove_item]["stock"] += removed_qty
    del cart[remove_item]
    print(f"{remove_item} removed from cart.")
else:
    print("Item not found in cart.")

# Display cart
print("\nShopping Cart:")
total = 0

for item, details in cart.items():
    subtotal = details["price"] * details["quantity"]
    total += subtotal

    print(f"{item} x{details['quantity']} = ₱{subtotal}")

print(f"\nTotal Price: ₱{total}")
```
